```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import seaborn as sns; sns.set()
from scipy.interpolate import griddata 
```

```python
# Gradient 1
g1_par = pd.read_csv(f'https://raw.githubusercontent.com/seaflow-uw/seaflow-sfl/refs/heads/master/curated/SCOPE_16_751.sfl', sep = '\t')

# Gradient 2
g2_par = pd.read_csv(f'https://raw.githubusercontent.com/seaflow-uw/seaflow-sfl/refs/heads/master/curated/MGL1704_740.sfl', sep = '\t')

# Gradient 3 
g3light_data = pd.read_csv('Grad3PAR.csv')
g3_par = pd.read_csv(f'https://raw.githubusercontent.com/seaflow-uw/seaflow-sfl/refs/heads/master/curated/KM1906_740.sfl', sep = '\t')
```

```python
# PAR data for gradients 1, 2 and 3
g1par_df = g1_par.dropna(subset = ['PAR'])
g1par_lat = g1par_df['LAT'].to_list()
g1par_data = g1par_df['PAR'].to_list()

g2par_df = g2_par.dropna(subset = ['PAR'])
g2par_lat = g2par_df['LAT'].to_list()
g2par_data = g2par_df['PAR'].to_list()

g3par_df = g3_par.dropna(subset = ['PAR'])
g3par_lat = g3par_df['LAT'].to_list()
g3par_data = g3par_df['PAR'].to_list()
```

```python
plt.xlabel('Lat')
plt.ylabel('PAR')
plt.scatter(g1par_lat, g1par_data, alpha = 1.0, color = 'crimson', label = 'Gradient 1')
plt.scatter(g2par_lat, g2par_data, alpha = 0.5, color = 'green', label = 'Gradient 2')
plt.scatter(g3par_lat, g3par_data, alpha = 0.25, color = 'pink', label = 'Gradient 3')
plt.legend()
```

```python
plt.title('Gradient 1 PAR')
plt.xlabel('Lat')
plt.ylabel('PAR')
plt.scatter(g1par_lat, g1par_data, alpha = 1.0, color = 'crimson', label = 'Gradient 1')
```

```python
plt.title('Gradient 2 PAR')
plt.xlabel('Lat')
plt.ylabel('PAR')
plt.scatter(g2par_lat, g2par_data, alpha = 0.5, color = 'green', label = 'Gradient 2')
```

```python
plt.title('Gradient 3 PAR')
plt.xlabel('Lat')
plt.ylabel('PAR')
plt.scatter(g3par_lat, g3par_data, alpha = 0.25, color = 'pink', label = 'Gradient 3')
```

```python
# creating scatter graph of light levels
plt.title('G3 Change of Light Levels')
plt.xlabel('Lat')
plt.ylabel('Surface PAR')

g3light_lat = g3light_data['lat'].to_list()
g3surface_par = g3light_data['surface_par'].to_list()

plt.scatter(g3light_lat, g3surface_par)
```
